```python
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/Alexander117161/Practica-Laboratorio-CD/master/df_sucio.csv')
df.head()
```

```python
resumen_estadístico = df.describe(include='all')
print("Resumen estadístico:\n", resumen_estadístico)
```

```python
# Calcular el porcentaje de valores faltantes por columna
porcentaje_faltantes = df.isnull().mean() * 100
print("\nPorcentaje de valores faltantes por columna:\n", porcentaje_faltantes)
```

```python
# Identificar filas duplicadas
duplicados = df.duplicated().sum()
print(f"\nNúmero de filas duplicadas: {duplicados}")
```

```python
# Tipos de datos de las columnas
tipos_datos = df.dtypes
print("\nTipos de datos de las columnas:\n", tipos_datos)
```

```python
# 1. Eliminación o Imputación de Valores Faltantes

# Revisar porcentaje de valores faltantes
porcentaje_faltantes = df.isnull().mean() * 100

# Decidir si eliminar o imputar en base a los valores faltantes
threshold = 50  # Si más del 50% de los valores faltan, eliminamos la columna
columnas_a_eliminar = porcentaje_faltantes[porcentaje_faltantes > threshold].index
df_limpio = df.drop(columns=columnas_a_eliminar)

# Para las columnas restantes, imputamos usando la mediana (opcional)
df_limpio.fillna(df_limpio.median(numeric_only=True), inplace=True)
df_limpio.head()
```

```python
# 2. Eliminación de Duplicados
df_limpio1 = df_limpio.drop_duplicates()
df_limpio1.head()
```

```python
# 3. Corrección de Tipos de Datos
# Listado de columnas que deberían ser numéricas y están como texto
columnas_a_convertir = ['latitude', 'longitude', 'temperature_celsius', 'temperature_fahrenheit',
                        'wind_mph', 'wind_kph', 'humidity', 'pressure_mb', 'precip_mm',
                        'feels_like_celsius', 'feels_like_fahrenheit', 'visibility_km', 'uv_index']

# Convertir a float
for columna in columnas_a_convertir:
    df_limpio1[columna] = pd.to_numeric(df_limpio[columna], errors='coerce')
# Convertir las fechas
df_limpio1['last_updated'] = pd.to_datetime(df_limpio['last_updated'], errors='coerce')
df_limpio1.head()
```

```python
# 4. Corrección de Valores "Invalid"
# Identificar y corregir valores como 'aaa', 'bbb', etc.
valores_invalidos = ['aaa', 'aaaa', 'aaaaa', 'bbb', 'cccc', 'xxxxx']  # Añadir más si es necesario
df_limpio1.replace(valores_invalidos, pd.NA, inplace=True)
df_limpio1.head()
```

```python
# Resumen estadístico antes de la limpieza
resumen_estadístico = df.describe(include='all')
print("Resumen estadístico antes de la limpieza:\n", resumen_estadístico)
```

```python
# Porcentaje de valores faltantes por columna antes de la limpieza
porcentaje_faltantes = df.isnull().mean() * 100
print("\nPorcentaje de valores faltantes por columna antes de la limpieza:\n", porcentaje_faltantes)
```

```python
# Número de filas duplicadas antes de la limpieza
duplicados = df.duplicated().sum()
print(f"\nNúmero de filas duplicadas antes de la limpieza: {duplicados}")
```

```python
# Tipos de datos antes de la limpieza
tipos_datos = df.dtypes
print("\nTipos de datos antes de la limpieza:\n", tipos_datos)
```

```python
# Resumen estadístico después de la limpieza
resumen_final = df_limpio1.describe(include='all')
print("Resumen estadístico después de la limpieza:\n", resumen_final)
```

```python
# Tipos de datos después de la limpieza
tipos_datos_despues = df_limpio1.dtypes
print("\nTipos de datos después de la limpieza:\n", tipos_datos_despues)
```

```python
# Porcentaje de valores faltantes por columna después de la limpieza
porcentaje_faltantes_despues = df_limpio.isnull().mean() * 100
print("\nPorcentaje de valores faltantes por columna después de la limpieza:\n", porcentaje_faltantes_despues)
```

```python
# Comprobar si aún existen valores inválidos
valores_invalidos = ['aaa', 'aaaa', 'bbb', 'cccc', 'xxxxx']  # Añadir más si es necesario
df_invalidos = df_limpio[df_limpio.isin(valores_invalidos).any(axis=1)]
print(f"\nFilas con valores inválidos después de la limpieza: {df_invalidos.shape[0]}")
```
